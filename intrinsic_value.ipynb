# Intrinsic Value - Manual DCF Method - Value as of 06-26-2021

## Variables:

 * Free Cash Flow (FCF) = Operating cash flow - Total CapEx (more conservative because it includes both growth and maintenance CapEx)

 * FCF Growth Rate 10 years (g) - 10-year average or make a judgement call what is a realistic growth rate to continue into the future
 
 * Personal Required Rate of Return (r) - use S&P historical avg return or specify your own requred rate

 * Long Term Growth Rate (lgr) - discretionary rate for earnings to grow indefinitely (recommended 2%-3% to avoid overly optimistic valuation) past ten years


* DFCFn = Discounted Free Cash Flow for year n
* FCFn = FV Free Cash Flow for year n
* r = Discount Factor for year n
* g = Growth Rate for FCF for year n
* lgr = Long Term Growth Rate
* n = Number of Periods

## Calculate FCF Discounted Cash Flow for n years:

* FCFn = BYFCF * (1+g)**n
* DFCFn = FCFn / (1+r)**n

## Calculate terminal (perpetual) value after n years:

* DPCF = (BYFCF * ((1+g)**(n+1))) * (1+lgr) / (r-lgr) * (1/((1+r)**(n+1)))

## Calculate Intrinsic value:

* Intrinsic Value = DFCF + DPCF

* Intrinsic Value per Share = Intrinsic Value / Average Weighted Shares Outstanding Diluted

```python
# Specify variables for DCF model

n = 10 # period ten years
r = 0.10 # discretionary factor based on the riskiness of investment
g = growth['freeCashFlowGrowth'].mean() # growth of earnings first ten years (average FCF)
lgr = 0.02 # long term growth rate to last indefinitely 
BYFCF = (quarterly_cash['freeCashFlow'][0]) * 4 # base year FCF annualized from the latest Q
shares_outstanding = quarterly_value['numberOfShares'][0]
```

```python
# FCF future value for period ten years
FCF  = BYFCF * ((1+g)**n) 

# Calculate discounted free cash flow for period of ten years
DFCF = FCF / ((1+r)**n)
DFCF.round(2)
```

```python
# Calculate terminal (perpetual) value past ten years
DPCF = (BYFCF * ((1+g)**(n+1))) * (1+lgr) / (r-lgr) * (1/((1+r)**(n+1)))
DPCF.round(2)
```

```python
# Calculate intrinsic value
intrinsic_value = DFCF + DPCF
intrinsic_value.round(2)
```

```python
# Calculate intrinsic value per share
intrinsic_value_per_share = intrinsic_value / shares_outstanding
intrinsic_value_per_share.round(2)
```

```python
# Calculate margin of safety
market_cap = quarterly_value['marketCapitalization'][0]
market_cap_per_share = market_cap / shares_outstanding

print(f'Current Market Value = {market_cap_per_share:.2f}')
print(f'Intrinsic Value = {intrinsic_value_per_share:.2f}')

# Calculate current margin
margin = (market_cap_per_share / intrinsic_value_per_share)
print(f'Margin = {margin:.2f}')

# Specify margin of safety
margin_of_safety = 0.70

if margin <= margin_of_safety:

    print('Seems like a viable investment option!')

else:

    print('Insufficient margin, keep searching!')
```

# FMP Intrinsic Value - [FMP API Calculation](https://financialmodelingprep.com/developer/docs/dcf-formula) - Value as of 09-21-2021

```python
url = f'https://financialmodelingprep.com/api/v3/discounted-cash-flow/{ticker}?&apikey={FMP_API_KEY}&limit=10'
dcf = requests.get(url).json()
dcf = pd.DataFrame(dcf)
dcf = dcf.rename(columns={'Stock Price':'stock_price'})
dcf['date'] = pd.to_datetime(dcf['date'])
dcf = dcf.set_index('date')
dcf['margin_fmp'] = dcf['stock_price'] / dcf['dcf']
pd.options.display.float_format = '{:.2f}'.format
dcf
```

```python
margin_of_safety = 0.70

if len(dcf[dcf['margin_fmp'] <= margin_of_safety]):

    print('Seems like a viable investment option!')

else:

    print('Insufficient margin, keep searching!')
```

# Intrinsic Value - Manual Simple Method (Value as of 09-26-2020)
---
## Variables:

 * Free Cash Flow (FCF) = Operating cash flow - Total CapEx (more conservative because it includes both growth and maintenance CapEx)

 * FCF Growth Rate 10 years (g) - 10-year averageg or make a judgement call what is realistic growth rate to continue into the future
 
 * Personal Required Rate of Return (r) - use S&P historical avg return or specify your own requred rate

 * Long Term Growth Rate (lgr) - discretionary rate for earnings to grow indefinitely (recommended 2%-3% to avoid overly optimistic valuation)


* DFCFn = Discounted Free Cash Flow for year n
* FCFn = Free Cash Flow for year n
* r = Discount Factor for year n
* g = Growth Rate for FCF for year n
* lgr = Long Term Growth Rate
* n = Number of Periods
---
## Calculate FCF Discounted Cash Flow for 10 years:

*  DFCFn = FCFn / (1+r)
---
## Calculate terminal value after ten years:

*  DPCF = ((BYFCF*(1+g)**11) * (1+g) / r-lgr) * (1/(1+r)**11)
---
## Calculate Intrinsic value:

*  Intrinsic Value = DFCF + DPCF

*  Intrinsic Value per Share = Intrinsic Value / Average Weighted Shares Outstanding Diluted

```python
# Specify variables for DCF model

n = 10 # period ten years
BYFCF = cash['freeCashFlow'].loc['2020'] # base year FCF
FCF  = BYFCF*(1+r)**n # FCF future value after reriod ten years
g = growth['freeCashFlowGrowth'].mean() # growth of earnings first ten years (average FCF)
lgr = 0.03 # long term growth rate to last indefinitely 
r = 0.10 # discretionary factor based on the riskiness of investment

shares_outstanding = income['weightedAverageShsOutDil'].loc['2020']
```

```python
# Calculate discounted free cash flow for period of ten years
DFCF = FCF / (1+r)
DFCF
```

```python
# Calculate terminal value past ten years
DPCF = ((BYFCF*(1+g)**11) * (1+g) / r-lgr) * (1/(1+r)**11)
DPCF
```

```python
# Calculate intrinsic value
intrinsic_value = DFCF + DPCF
intrinsic_value
```

```python
# Calculate intrinsic value per share
intrinsic_value_per_share = intrinsic_value / shares_outstanding
intrinsic_value_per_share
```

```python
# Calculate margin of safety
market_cap = key_metrics['marketCap'].loc['2020']
market_cap_per_share = market_cap / shares_outstanding
print(f'Current Market Value = {market_cap_per_share}')
print(f'Intrinsic Value = {intrinsic_value_per_share}')

# Calculate current margin
margin = market_cap_per_share / intrinsic_value_per_share
print(f'Margin = {margin}')

# Specify margin of safety
margin_of_safety = 0.70

if len(margin) <= margin_of_safety:

    print('Seems like a viable investment option!')

else:

    print('Insufficient margin, keep searching!')
```

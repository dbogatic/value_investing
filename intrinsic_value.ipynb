# Intrinsic Value - [FMP API Calculation](https://financialmodelingprep.com/developer/docs/dcf-formula)

```python
url = f'https://financialmodelingprep.com/api/v3/discounted-cash-flow/{ticker}?&apikey={FMP_API_KEY}&limit=10'
dcf = requests.get(url).json()
dcf = pd.DataFrame(dcf)
dcf = dcf.rename(columns={'Stock Price':'stock_price'})
dcf['date'] = pd.to_datetime(dcf['date'])
dcf = dcf.set_index('date')
dcf['margin'] = dcf['stock_price'] / dcf['dcf']
pd.options.display.float_format = '{:.2f}'.format
dcf
```

```python
margin_of_safety = 0.70

if len(dcf[dcf['margin'] <= margin_of_safety]):

    print('Seems like a viable investment option')

else:

    print('Not good, keep searching!')
```

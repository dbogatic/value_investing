```python
import yfinance as yf
from yahoofinancials import YahooFinancials
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
# Specify stock ticker
ticker = 'AAPL'

# Need to call after specifying stock ticker for pulling financial statements from yahoofinancials 
yahoo_financials = YahooFinancials(ticker)
```

```python
# Obtain income statements
income = yahoo_financials.get_financial_stmts('annual', 'income')
income_one = pd.DataFrame(income['incomeStatementHistory'][ticker][0])
income_two = pd.DataFrame(income['incomeStatementHistory'][ticker][1])
income_three = pd.DataFrame(income['incomeStatementHistory'][ticker][2])
income_four = pd.DataFrame(income['incomeStatementHistory'][ticker][3])
income_combined = pd.concat([income_one,income_two,income_three,income_four], axis=1).astype('float')
pd.options.display.float_format = "{:,.2f}".format
income_combined
```

```python
# Obtain balance sheet
balance = yahoo_financials.get_financial_stmts('annual', 'balance')
balance_one = pd.DataFrame(balance['balanceSheetHistory'][ticker][0])
balance_two = pd.DataFrame(balance['balanceSheetHistory'][ticker][1])
balance_three = pd.DataFrame(balance['balanceSheetHistory'][ticker][2])
balance_four = pd.DataFrame(balance['balanceSheetHistory'][ticker][3])
balance_combined = pd.concat([balance_one,balance_two,balance_three,balance_four], axis=1).astype('float')
pd.options.display.float_format = "{:,.2f}".format
balance_combined
```

```python
# Obtain annual cash-flow statements
cash = yahoo_financials.get_financial_stmts('annual', 'cash')
cash_one = pd.DataFrame(cash['cashflowStatementHistory'][ticker][0])
cash_two = pd.DataFrame(cash['cashflowStatementHistory'][ticker][1])
cash_three = pd.DataFrame(cash['cashflowStatementHistory'][ticker][2])
cash_four = pd.DataFrame(cash['cashflowStatementHistory'][ticker][3])
cash_combined = pd.concat([cash_one,cash_two,cash_three,cash_four], axis=1).astype('float')
pd.options.display.float_format = "{:,.2f}".format
cash_combined
```
